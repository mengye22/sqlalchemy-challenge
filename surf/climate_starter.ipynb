```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
from scipy import stats
from scipy import mean
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine,func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine,reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the last date in the table
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first().date
latest_year = dt.datetime.strptime(latest_date, '%Y-%m-%d') - dt.timedelta(days=365)
latest_year = latest_year.strftime("%Y-%m-%d")
# Perform a query to retrieve the data and precipitation scores
# Sort the dataframe by date
precipitation_date = session.query(Measurement.date,Measurement.prcp).\
    filter(Measurement.date >= latest_year).\
    order_by(Measurement.date).all()
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(precipitation_date)
prcp_df.set_index('date',inplace=True)

# Use Pandas Plotting with Matplotlib to plot the data
fig,ax = plt.subplots()
prcp_df.plot(ax=ax,figsize=(15,10))
ax.set_xlabel("date")
ax.set_ylabel("precipitation")
ax.set_title(f"Precipitation between {latest_year} to {latest_date}")
fig.tight_layout()
plt.savefig("Images/Precitation.png")
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcp_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station.id).distinct().count()
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_sum = session.query(Measurement.station,func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
station_sum
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
sel = [Measurement.station,
      func.min(Measurement.tobs),
      func.max(Measurement.tobs),
      func.avg(Measurement.tobs)]
temp = session.query(*sel).filter(Measurement.station == station_sum[0][0]).all()
temp
```

```python
print(f"The lowest Temperature for {temp[0][0]} is {temp[0][1]}.")
print(f"The highest Temperature for {temp[0][0]} is {temp[0][2]}.")
print(f"The average Temperature for {temp[0][0]} is {round(temp[0][3],2)}.")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and 
# plot the results as a histogram

# 
latest_temp = session.query(Measurement.date, Measurement.tobs).\
    filter(Measurement.date >= latest_year).\
    filter(Measurement.station == temp[0][0]).\
    order_by(Measurement.date).all()
# 
temp_df = pd.DataFrame(latest_temp)
temp_df = temp_df.set_index('date')
temp_df.plot(kind='hist',bins=12,figsize=(9,6))
plt.title(f"Lastest Year Temperature of {temp[0][0]}")
plt.xlabel("Temperature")
fig.tight_layout()
plt.savefig("Images/highest_count_station_temp.png")
plt.show()
```

```python
# average temperature in June at all stations across all available years in the dataset
june_str = "06"
# list of June temperature 
june_temp = session.query(Measurement.tobs).\
    filter(func.strftime("%m", Measurement.date) == june_str).all()
# average temperature in June 
june_avg = np.mean(june_temp)
# average temperature in December at all stations across all available years in the dataset

dec_str = "12"
# list of June temperature 
dec_temp = session.query(Measurement.tobs).\
    filter(func.strftime("%m", Measurement.date) == dec_str).all()
# average temperature in December
dec_avg = np.mean(dec_temp)

# print out average
print(f"The avgerage temperature for June is: {june_avg}")
print(f"The avgerage temperature for December is: {dec_avg}")
```

```python
# apple T-test on June and December datasets
stats.ttest_ind(june_temp,dec_temp)
```

#### Discover
1. Independent T-test for the datasets, since there is temperature between June and December is not dependent on each other.
2. Since p-value is smaller that 0.5, this means the test is statistically significant.

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
start_trip = '2017-06-01'
end_trip = '2017-06-13'
my_trip_temp = calc_temps(start_trip,end_trip)
my_trip_temp
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
tmin = my_trip_temp[0][0]
tavg = my_trip_temp[0][1]
tmax = my_trip_temp[0][2]

fig,ax = plt.subplots(figsize=(4,8))
ax.bar(0,tavg,yerr=(tmax-tmin),width=0.4)
ax.set_title('Trip Avg Temp')
ax.set_ylabel('Temperature (F)')
ax.set_xticks([])
plt.savefig("Images/trip_avg_temp.png")
plt.show()
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates 
# using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the 
# station, name, latitude, longitude, and elevation

sel = [Station.station,
       func.sum(Measurement.prcp),
       Station.name,
       Station.latitude,
       Station.longitude,
       Station.elevation
      ]

trip_datas = session.query(*sel).\
    filter(Station.station == Measurement.station).\
    filter(Measurement.date >= start_trip).\
    filter(Measurement.date <= end_trip).\
    group_by(Station.station).\
    order_by(func.sum(Measurement.prcp).desc()).all()

trip_df = pd.DataFrame(trip_datas,columns = ["Station",
                                            "Precipitation",
                                            "Name",
                                            "Latitude",
                                            "Longitude",
                                            "Elevation"])
trip_df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
start_date = '2017-03-17'
end_date = '2017-03-31'
start = dt.datetime.strptime(start_date, '%Y-%m-%d')
end = dt.datetime.strptime(end_date, '%Y-%m-%d')

# Use the start and end date to create a range of dates
trip_range = (dt.datetime.strptime(end_date,'%Y-%m-%d') - dt.datetime.strptime(start_date,'%Y-%m-%d')).days + 1

trip_dates = []
while (start <= end):
    trip_dates.append(dt.datetime.strftime(start,"%m-%d"))
    start = start + dt.timedelta(days=1)

# # Loop through the list of %m-%d strings and calculate the normals for each date
normal = []
for date in range(0,trip_range):
    normal.append(list(np.ravel(daily_normals(trip_dates[date]))))
    
normal
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
normal_df = pd.DataFrame(normal,columns=['TMIN','TAVG','TMAX'])
normal_df["Date"] = trip_dates
normal_df = normal_df.set_index("Date")
normal_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
normal_df.plot(kind="area",stacked=False,figsize=(10,6))
plt.xticks(rotation=45)
plt.title("Trip Daily Normal")
plt.ylabel("Temperature(F)")
plt.savefig("Images/trip_daily_normal.png")
plt.show()
```
